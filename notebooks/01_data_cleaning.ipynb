# Data Cleaning & Quality Check
Verify the Superstore dataset is clean and document any anomalies before EDA.

```python
import sys
sys.path.append("..")

import pandas as pd
import plotly.express as px
from src.data_loader import load_data, clean_data

df_raw = load_data("../data/superstore.csv")
df = clean_data(df_raw)
```

```python
print("Shape:", df.shape)
print("\n--- Sales ---")
print(df["sales"].describe())
print("\n--- Profit ---")
print(df["profit"].describe())
print("\n--- Discount ---")
print(df["discount"].describe())
```

```python
out_of_range = df[(df["discount"] < 0) | (df["discount"] > 1)]
print(f"Discount out of [0, 1] range: {len(out_of_range)} rows")
```

```python
anomalies = df[(df["profit_margin"] < -50) | (df["profit_margin"] > 100)]
print(f"Profit margin anomalies (<-50% or >100%): {len(anomalies)} rows")
print(anomalies[["product_name", "sales", "profit", "profit_margin", "discount"]].head(10))
```
